```python
#clean test data
test_df = pd.read_csv('test.csv',header = 0)
test_df['Gender'] = test_df['Sex'].map({'female': 0, 'male': 1}).astype(int)

#replace null embarked values with most common embarked values and convert to integer
if len(test_df.Embarked[test_df.Embarked.isnull()]) > 0:
    test_df.Embarked[test_df.Embarked.isnull()] = test_df.Embarked.dropna().mode().values
test_df['Port'] = test_df.Embarked.map({'C': 0, 'S': 1, 'Q': 2}).astype(int)

#fill in missing ages with avg age for gender and class
test_df['AgeFill'] = test_df['Age']
median_ages = np.zeros((2, 3))
for r in range(2):
    for c in range(3):
        median_ages[r, c] = test_df[(test_df['Gender'] == r) & (test_df['Pclass'] == c + 1)]['Age'].dropna().median()
for r in range(2):
    for c in range(3):
        test_df.loc[(test_df.Age.isnull()) & (test_df.Gender == r) & (test_df.Pclass == c + 1), 'AgeFill'] = median_ages[r, c]
        
#add ageisnull, familysize, age*class
test_df['AgeIsNull'] = pd.isnull(test_df.Age).astype(int)
test_df['FamilySize'] = (test_df.SibSp + test_df.Parch)
test_df['Age*Class'] = (test_df.AgeFill * test_df.Pclass)
        
#fill in missing fare values
if len(test_df.Fare[test_df.Fare.isnull()]) > 0:
    median_fare = np.zeros(3)
    for c in range(3):
        median_fare[c] = test_df[test_df.Pclass == c + 1]['Fare'].dropna().median()
    for c in range(3):
        test_df.loc[(test_df.Fare.isnull()) & (test_df.Pclass == c + 1), 'Fare'] = median_fare[c]
        
#save passenger ids
ids = test_df.PassengerId.values

#drop unneeded columns and string columns
test_df = test_df.drop(['PassengerId', 'Name', 'Sex', 'Cabin', 'Ticket', 'Embarked', 'Age'], axis = 1)

test_data = test_df.values


test_df.info()
```

```python
# Machine learning with scikit-learn package

from sklearn.ensemble import RandomForestClassifier

print 'Training...'
forest = RandomForestClassifier(n_estimators = 100)
forest = forest.fit(train_data[:, 1:], train_data[:, 0])

print 'predicting...'
output = forest.predict(test_data).astype(int)

predictions_file = open('myfirstforest.csv', 'wb')
open_file_object = csv.writer(predictions_file)
open_file_object.writerow(['PassengerId', 'Survived'])
open_file_object.writerows(zip(ids, output))
predictions_file.close()
print 'Done'
```
